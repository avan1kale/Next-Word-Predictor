```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
with open("data.txt", "r", encoding="utf-8") as file:
    text = file.read()

print(text[:200])  # first 200 characters
```

```python
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
for sentence in text.split('.'):
    print(sentence.strip())
```

```python
input_sequence=[]
for sentence in text.split('.'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence.strip()])[0]
    print(tokenized_sentence)

    for i in range(1, len(tokenized_sentence)):
        input_sequence.append(tokenized_sentence[:i+1])
```

```python
input_sequence
```

```python
#need to apply padding to the input sequences so that they are all of the same length. This is important for training the LSTM model, as it requires inputs of uniform size.
maxLen=max([len(x) for x in input_sequence])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequence, maxlen=maxLen, padding='pre')
```

```python
padded_input_sequences
```

```python
# now our dataset is ready for training. The next steps would typically involve splitting the data into features and labels, defining the LSTM model architecture.
# the last number in each padded_input_sequences is the label (the next word to predict), and the rest are the features (the previous words). So we can split the data accordingly.
X=padded_input_sequences[:,:-1]
y=padded_input_sequences[:,-1]
```
