```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
with open("data.txt", "r", encoding="utf-8") as file:
    text = file.read()

print(text[:200])  # first 200 characters
```

```python
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
for sentence in text.split('.'):
    print(sentence.strip())
```

```python
input_sequence=[]
for sentence in text.split('.'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence.strip()])[0]
    print(tokenized_sentence)

    for i in range(1, len(tokenized_sentence)):
        input_sequence.append(tokenized_sentence[:i+1])
```

```python
input_sequence
```

```python
#need to apply padding to the input sequences so that they are all of the same length. This is important for training the LSTM model, as it requires inputs of uniform size.
maxLen=max([len(x) for x in input_sequence])
```

```python
maxLen
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequence, maxlen=maxLen, padding='pre')
```

```python
padded_input_sequences
```

```python
# now our dataset is ready for training. The next steps would typically involve splitting the data into features and labels, defining the LSTM model architecture.
# the last number in each padded_input_sequences is the label (the next word to predict), and the rest are the features (the previous words). So we can split the data accordingly.
X=padded_input_sequences[:,:-1]
y=padded_input_sequences[:,-1]
```

```python
X
```

```python
y
```

```python
#take this as a classification problem, we need to one-hot encode the labels. The number of classes will be equal to the size of the vocabulary (i.e., the number of unique words in the text).
#the output of the one-hot encoding will be a binary matrix representation of the labels, where each row corresponds to a label and each column corresponds to a class (word in the vocabulary). The value at a given row and column will be 1 if the label corresponds to that class, and 0 otherwise.
from tensorflow.keras.utils import to_categorical
y=to_categorical(y, num_classes=len(tokenizer.word_index)+1)
#we add 1 to the number of classes because the tokenizer's word index starts from 1, and we need to account for the 0 index as well during onehot encoding.
```

```python
y.shape
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

model = Sequential()
model.add(Embedding(8932,100,input_length =105, input_shape=(105,)))
model.add(LSTM(150))
model.add(Dense(8932,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X.shape
```

```python
model.summary()
#input dimension=100
# in embedding layer- for every word in a sentence, we will have a 100-dimensional vector representation. The input_length=105 means that each input sequence will be of length 105 (the maximum length of the sequences after padding).
#hence param for embedding layer= 8932*100=893300 since there are 8932 unique words in the vocabulary and each word is represented by a 100-dimensional vector.

#in lstm layer, the timesteps will be 105 (the length of the input sequences), and the number of features will be 100 (the dimensionality of the word embeddings).
#the output will be of (1,150) since we have 150 units in the LSTM layer. we know diemensionality of H(t) is equal to number of nodes in LSTM layer.

#in the dense layer, the input is the 150-dimensional output from the LSTM layer, and the output is a 8932-dimensional vector (the size of the vocabulary). The softmax activation function is used to convert the output into a probability distribution over the vocabulary.
```

```python
model.fit(X, y, epochs=60, verbose=1)
```

```python
word2="enough"
for i in range(5):
    
    token_text=tokenizer.texts_to_sequences([word2])[0]
    pad_text=pad_sequences([token_text], maxlen=105, padding='pre')
    predicted_word_index=np.argmax(model.predict(pad_text))
    predicted_word=tokenizer.index_word[predicted_word_index]
    print(predicted_word)
    word2+= " " + predicted_word
```
